```python
import librosa
import soundfile as sf


def resample_audio(input_file, output_file, target_sr):
    y, sr = librosa.load(input_file, sr=None)
    y_resampled = librosa.resample(y, orig_sr=sr, target_sr=target_sr)
    sf.write(output_file, y_resampled, target_sr)

dir_in = "audio"
for faudio in os.listdir(dir_in):
    if faudio.split('.')[-1]=='wav':
        wav_file = os.path.join(dir_in, faudio)
        wav_fileout = os.path.join(dir_in, faudio.replace('_', ''))
        
        resample_audio(wav_file, wav_fileout, 16000)
```

```python
embeddings_dataset = load_dataset("Matthijs/cmu-arctic-xvectors", split="validation")
vocoder = SpeechT5HifiGan.from_pretrained("microsoft/speecht5_hifigan")
```

```python
speaker_embeddings = torch.tensor(embeddings_dataset[7306]["xvector"]).unsqueeze(0)
```

```python
processor = SpeechT5Processor.from_pretrained("microsoft/speecht5_vc")
model = SpeechT5ForSpeechToSpeech.from_pretrained("microsoft/speecht5_vc")
```

```python
dataset = load_dataset("hf-internal-testing/librispeech_asr_demo", "clean", split="validation")
example = dataset[40]
```

```python
Audio(example["audio"]["array"], rate=16000)
```

```python
sampling_rate = dataset.features["audio"].sampling_rate
inputs = processor(audio=example["audio"]["array"], sampling_rate=sampling_rate, return_tensors="pt")
speech = model.generate_speech(inputs["input_values"], speaker_embeddings, vocoder=vocoder)
```

```python
Audio(speech, rate=16000)
```
